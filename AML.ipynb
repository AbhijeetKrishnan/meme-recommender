```
import pandas as pd
import glob
```

```
f=sorted(glob.glob('/content/gdrive/My Drive/memes/*'))
f1=f[:1000]   #Take the first 1000
```

```
from keras.preprocessing import image
from skimage import color
from skimage.feature import hog
```

```
print(f1)
```

```
ls=[]
for x in f1:
  img = image.load_img(x,target_size=(224, 224))
  arr=image.img_to_array(img)
  data_gray =color.rgb2gray(arr)
  ls.append(data_gray)
```

```
len(ls)
```

```
ppc = 32
hog_images = []
hog_features = []
for x in ls:
    feature_vec,image = hog(x, orientations=8, pixels_per_cell=(ppc,ppc),cells_per_block=(4, 4),block_norm= 'L2',visualise=True)
    hog_images.append(image)
    print(image)
    hog_features.append(feature_vec)
```

```
import numpy as np
hog_features=np.array(hog_features)
```

```
hog_features.shape
```

```
from sklearn.cluster import KMeans
```

```
kmeans = KMeans(n_clusters=3, random_state=0)
cl=kmeans.fit_predict(hog_features)
print(cl)
```

```
import numpy as np
np.savetxt("/content/gdrive/My Drive/clusters_hog.csv", kmeans, delimiter=",")
```
