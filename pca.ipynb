```python
import glob
from keras.preprocessing import image
```

```python
f = sorted(glob.glob("data/memes/*"))
f = f[:1000]  # Take the first 1000
print(f)
```

```python
import numpy as np
from skimage import color

ls = []
faces = np.zeros((0, 10000))
for x in f:
    img = image.load_img(x, target_size=(100, 100))
    arr = image.img_to_array(img)
    arr = color.rgb2gray(arr)
    face = arr.flatten()
    faces = np.vstack([faces, face])
```

```python
print(faces)
```

```python
from sklearn.decomposition import PCA

# n_components=0.80 means it will return the Eigenvectors that have the 80% of the variation in the dataset
faces_pca = PCA(n_components=0.8)
faces_pca.fit(faces)
```

```python
components = faces_pca.transform(faces)
```

```python
print(components.shape)
print(type(components))
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=0).fit_predict(components)
print(kmeans)
```
